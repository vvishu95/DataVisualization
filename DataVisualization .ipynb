```python
# Now we will draw the graph according to given condition
```

```python
def rank_performance(stock_price):
    if stock_price <= 10:
        return "Poor"
    elif stock_price <= 50:
        return "Satisfactory"
    else:
        return "Steller"
```

```python
# Count of Poor, Satisfactory and Steller stock prices
```

```python
bb["Close"].apply(rank_performance).value_counts()
```

```python
# Now we are using Horizontal Bar Graph to show the result
```

```python
plt.style.use("ggplot")
bb["Close"].apply(rank_performance).value_counts().plot(kind = "barh")
```

```python
bb["Close"].mean()
```

```python
# Now we are using another condition
```

```python
def rank_performance(mean):
    if mean >= 30.486829397664266:
        return "Above Average"
    else:
        return "Below Average"
```

```python
# Count of Above average and Below Average
```

```python
bb["Close"].apply(rank_performance).value_counts()
```

```python
# Now we will show the result in Pie graph and will use dark backgound
```

```python
plt.style.use("dark_background")
bb["Close"].apply(rank_performance).value_counts().plot(kind = "pie", legend = "True")
```
